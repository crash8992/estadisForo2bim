```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal estándar
mu = 0
sigma = 1

# Nivel de confianza (área central)
confianza = 0.95

# Área en ambas colas
area_colas = 1 - confianza

# Área en cada cola
area_cola = area_colas / 2

# Valores críticos Z para las colas (simétricos)
z_critico = norm.ppf(1 - area_cola)

print(f"Área total en ambas colas: {area_colas:.4f}")
print(f"Área en cada cola: {area_cola:.4f}")
print(f"Valores críticos Z: ±{z_critico:.4f}")

# Crear rango
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Graficamos
plt.plot(x, y, label='Curva Normal')

# Sombrear cola izquierda
x_fill_left = np.linspace(mu - 4*sigma, -z_critico, 1000)
y_fill_left = norm.pdf(x_fill_left, mu, sigma)
plt.fill_between(x_fill_left, y_fill_left, color='red', alpha=0.5, label='Cola izquierda')

x_fill_right = np.linspace(z_critico, mu + 4*sigma, 1000)
y_fill_right = norm.pdf(x_fill_right, mu, sigma)
plt.fill_between(x_fill_right, y_fill_right, color='red', alpha=0.5, label='Cola derecha')

# valores críticos
plt.axvline(-z_critico, color='red', linestyle='--')
plt.axvline(z_critico, color='red', linestyle='--')

# Etiquetas y leyenda
plt.title(f'Área en ambas colas (total = {area_colas:.2%}) de la curva normal')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()


```
